## 01 | Visualize Cricket Data with Matplotlib and Pandas

The second example notebook shows how to load data from a file into a data frame and visualize a simple plot. Let's see if I can replicate this with my own data file - and add at least one new visualization in the process. 
This notebooks reads sample cricket data from `1-data/kaggle/IPL-2022.csv and plots it using Matplotlib using a minimally modified version of the code from the population example before.
We also use the df.columns function to get a list of the column names in the data - so we can use them in our code changes

```python
# Loads the relevant Python libraries
import matplotlib.pyplot as plt
import pandas

# Creates a dataframe from the CSV file 
df = pandas.read_csv('../1-data/kaggle/IPL-2022.csv')

## Get the column names
print(df.columns)

## Loads column data into relevant Series objects (x and y)
## Plots values of x and y
## Can use this to see if y is a function of x 
x=df['first_ings_wkts']
y=df['first_ings_score']
plt.plot(x,y)
plt.title("Wickets vs Score")
plt.xlabel('Wickets')
plt.ylabel('Score')

## Plot seems to indicate there is no clear relationship between them
plt.show()

```
