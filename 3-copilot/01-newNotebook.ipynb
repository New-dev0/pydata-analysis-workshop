# Import Required Libraries
Import the necessary libraries, including pandas for data manipulation and seaborn for data visualization.

```python
# Importing the necessary libraries
import pandas as pd
import seaborn as sns
```

# Load the Cricket Data File
Use pandas to load the cricket data file into a DataFrame.

```python
# Load the cricket data file
cricket_data = pd.read_csv('./../1-data/kaggle/IPL-2022.csv')

# Display the first few rows of the DataFrame
cricket_data.head()
```

# Data Overview
Perform basic data exploration to understand the structure and content of the cricket data.

```python
# Display the shape of the DataFrame
print(cricket_data.shape)

# Display the column names
print(cricket_data.columns)

# Display the data types of each column
print(cricket_data.dtypes)

# Display the summary statistics of the DataFrame
print(cricket_data.describe())

# Check for missing values
print(cricket_data.isnull().sum())

# Display the unique values in each column
for column in cricket_data.columns:
    print(f"{column}: {cricket_data[column].nunique()} unique values")
```

# Create Basic Seaborn Plots
Create basic Seaborn plots such as scatter plots, line plots, and histograms to visualize the cricket data.

```python
# Create Basic Seaborn Plots

# Scatter plot
sns.scatterplot(x='first_ings_score', y='first_ings_wkts', data=cricket_data)
```
